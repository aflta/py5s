<h2>Il metodo dei trapezi</h2>
<p>L'idea: approssimare l'area compresa fra una funzione $f$ e l'asse $x$ con l'area di $n$ trapezi<p>
<p>Area trapezio? Somma basi per altezza diviso due, ovvero altezza per semisomma basi</p>
<p>$\int_{a}^{b} f(x) \, \mathrm{d}x \simeq \frac{b-a}{n} \sum_{i=0}^{n-1} \frac{f_i(x)+f_{i+1}(x)}{2}$</p>

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return (x-3)*(x-5)*(x-7)+85

x = np.linspace(0, 10, 200)
y = f(x)
```

```python
a, b = 1, 8 # the left and right boundaries
N = 10 # the number of points
xint = np.linspace(a, b, N)
yint = f(xint)
```

```python
plt.plot(x, y, lw=2)
plt.axis([0, 9, 0, 140])
plt.fill_between(xint, 0, yint, facecolor='gray', alpha=0.4)
plt.text(0.5 * (a + b), 30,r"$\int_a^b f(x)dx$", horizontalalignment='center', fontsize=20);
```
